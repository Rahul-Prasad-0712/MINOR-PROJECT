```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,10)
```

```python
df = pd.read_csv("Suicides in India 2001-2012.csv")
df.head()
```

```python
df.shape
```

```python
df.groupby('State')['State'].agg('count')
```

```python
df1 = df.drop(['Type_code'],axis = 'columns')
df1.head()
```

## Data Cleaning

```python
df1.isnull().sum()
```

```python
df1['Type'].unique()
```

```python
df1.groupby('Type')['Type'].agg('count')
```

```python
len(df1.Type.unique())
```

```python
df1.location = df1.Type.apply(lambda x: x.strip())
type_stats = df1.groupby('Type')['Type'].agg('count').sort_values(ascending = False)
type_stats
```

```python
Type_stats_less_than_10 = type_stats[type_stats<=10]
```

```python
df1.Type = df1.Type.apply(lambda x: 'other' if x in Type_stats_less_than_10 else x)
```

```python
df1.Type.unique()
```

```python
len(df1.Type.unique())
```

```python
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df1['Gender'] = le.fit_transform(df1['Gender'])
df1['Age_group'] = le.fit_transform(df1['Age_group'])
df1['Type'] = le.fit_transform(df1['Type'])
```

```python
df1.head()
```

## Model Building

```python
dummies = pd.get_dummies(df1.State)
dummies.head(3)
```

```python
dummies.drop('Total (All India)',axis = 'columns')
```

```python
dummies.drop('Total (States)',axis = 'columns')
```

```python
dummies.drop('Total (Uts)',axis = 'columns')
```

```python
df2 = pd.concat([df1,dummies],axis = 'columns')
df2.head()
```

```python
df2.shape
```

```python
df3 = df2.drop('State',axis = 'columns')
```

```python
X = df3.drop('Total',axis = 'columns')
X.head()
```

```python
y = df3.Total
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5, test_size = 0.2,random_state = 0)

cross_val_score(LinearRegression(),X,y,cv =cv)
```

## Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression':{
            'model': LinearRegression(),
            'params':{
                'normalize': [True, False]
                
            }   
        },
        'lasso' : {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection':['random','cyclic']
            }
        },
        
        'random_forest':{
            'model' : RandomForestRegressor(),
            'params':{
                "n_estimators"      : [10,20,30],
                "max_features"      : ["auto", "sqrt", "log2"],
                "min_samples_split" : [2,4,8],
                "bootstrap": [True, False],
                
            }
        },
        
        
    }
    scores = []
    cv = ShuffleSplit(n_splits = 5,test_size = 0.2,random_state = 0)
    for algo_name,config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv = cv, return_train_score = False)
        gs.fit(X,y)
        scores.append({
            'model':algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns = ['model','best_score','best_params'])
find_best_model_using_gridsearchcv(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
rfr.score(X_train,y_train)
```

```python
def predict_suicide(State,Year,Type,Gender,Age_group):
    state_index = np.where(X.columns == State)[0]
    
    x = np.zeros(len(X.columns))
    x[0] = Year
    x[1] = Type
    x[2] = Gender
    x[3] = Age_group
    
    if state_index >= 0:
        x[state_index] = 1
        
    return rfr.predict([x])[0]
```

```python
predict_suicide("Chattisgarh",2015,5,1, 3)
```
